```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Load dataset
df = sns.load_dataset('iris')

print("First 5 rows:\n", df.head())

# --------------------------------------
# 1. Summary statistics grouped by categorical variable
# --------------------------------------

# Group by species and calculate statistics
grouped_stats = df.groupby('species').agg(['mean', 'median', 'min', 'max', 'std'])

print("\nSummary Statistics Grouped by Species:\n")
print(grouped_stats)

# Create list of numeric values for each category (example: sepal_length)
grouped_list = df.groupby('species')['sepal_length'].apply(list)

print("\nList of Sepal Length values for each species:\n")
print(grouped_list)

# --------------------------------------
# 2. Statistical details for each species
# --------------------------------------

species_list = df['species'].unique()

for sp in species_list:
    print(f"\nStatistics for {sp}:\n")
    subset = df[df['species'] == sp]

    print(subset.describe())   # includes mean, std, percentiles
```
