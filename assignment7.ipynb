```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

# Download once
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

text = "Natural Language Processing is a field of Artificial Intelligence."

# Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# POS Tagging
print("POS Tags:", pos_tag(tokens))

# Stopword Removal
stop = set(stopwords.words('english'))
words = [w for w in tokens if w.lower() not in stop]
print("After Stopword Removal:", words)

# Stemming
ps = PorterStemmer()
print("Stemming:", [ps.stem(w) for w in words])

# Lemmatization
lm = WordNetLemmatizer()
print("Lemmatization:", [lm.lemmatize(w) for w in words])

# TF-IDF
docs = [
    text,
    "Machine learning is part of AI",
    "NLP helps computers understand language"
]

tfidf = TfidfVectorizer()
result = tfidf.fit_transform(docs)

print("TF-IDF Matrix:\n", result.toarray())
```
