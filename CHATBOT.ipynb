```python
def chatbot():
  print("hello chat-bot here")
  print("how may i help you?")
  while True:
    user = input("you : ").lower()
    if user == "hello" or user =="hi":
      print("Bot: heyy, how may i help you?")

    elif "product" in user:
      print("Bot: Our products are available online.")

    elif "price" in user:
      print("Bot:Product price starts from Rs.500")

    elif "delivery" in user:
      print("Bot:Delivery can take 3-5 days")

    elif "bye" in user:
      print("Bot:bye have a good day")
      break

    else:
      print("Bot: Sorry, I don't understand.")

chatbot()
```
